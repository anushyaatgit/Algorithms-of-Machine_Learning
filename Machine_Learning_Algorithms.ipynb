```python
#3 Polynomial Regression

#Read dataset

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data_set=pd.read_csv("C:\\Users\\Renji\\Desktop\\Anu\\M L ASGNT DATAS\\WeatherDataP.csv")
print(data_set)
df=pd.DataFrame(data_set)
df
```

```python
#Info About the Data

df.info()
```

```python
df.isnull().sum()
```

```python
#sorting is done on row labels in ascending order.

df.sort_values(["Pressure (millibars)","Humidity"])
```

```python
#extract the dependent and independent variables from the given dataset

x=data_set.iloc[:,:-1].values #Independent variable
y=data_set.iloc[:,1].values #Dependent variable

x1=pd.DataFrame(x)
print(" Pressure (millibars)")
print(x1)

y1=pd.DataFrame(y)
print("Humidity")
print(y1)
```

```python
#Fitting the Linear Regression to the dataset

from sklearn.linear_model import LinearRegression
lin_regs= LinearRegression() 
lin_regs.fit(x,y)

#Fitting the Polynomial regression to the dataset

from sklearn.preprocessing import PolynomialFeatures
#The parameter value(degree= 2) depends on our choice. We can choose it according to our Polynomial features.
poly_regs= PolynomialFeatures(degree= 2)
#we are converting our feature matrix into polynomial feature matrix, and then fitting it to the Polynomial regression
x_poly= poly_regs.fit_transform(x) 
lin_reg_2 =LinearRegression() 
lin_reg_2.fit(x_poly, y)
print("done")

#Visualizing the result for Linear regression:
#Visulaizing the result for Linear Regression model

plt.scatter(x,y,color="blue") 
plt.plot(x,lin_regs.predict(x), color="red")
plt.title("Bluff detection model(Linear Regression)")
plt.xlabel("Pressure (millibars)")
plt.ylabel("Humidity")
plt.show()
```

```python
#Visulaizing the result for Polynomial Regression

plt.scatter(x,y,color="blue")
plt.plot(x, lin_reg_2.predict(poly_regs.fit_transform(x)), color="red")
plt.title("Bluff detection model(Polynomial Regression)")
plt.xlabel("Pressure (millibars)") 
plt.ylabel("Humidity")
plt.show()
```

```python
#Predicting the final result with the Linear Regression model:
lin_pred = lin_regs.predict([[1012.5]]) 
print(lin_pred)
#Predicting the final result with the Polynomial Regression model:
poly_pred = lin_reg_2.predict(poly_regs.fit_transform([[1012.5]])) 
print(poly_pred)
```

```python
#4 Logistic Regression

#Read dataset

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data_set=pd.read_csv("C:\\Users\\Renji\\Desktop\\Anu\\M L ASGNT DATAS\\WeatherAlbury.csv")
print(data_set)
df=pd.DataFrame(data_set)
df
```

```python
#Summarizing Data

df.describe()
```

```python
#Info About the Data

df.info()
```

```python
df.isnull().sum()
```

```python
df.RainToday = [1 if each == 'Yes' else 0 for each in df.RainToday]
df.RainTomorrow = [1 if each == 'Yes' else 0 for each in df.RainTomorrow]
df
```

```python
df.fillna(df.mean(), inplace=True)
df
```

```python
df.isnull().sum()
```

```python
#split dataset in features and target variable
feature_cols = ["MinTemp","MaxTemp","Rainfall","Humidity9am","Humidity3pm","Pressure9am","Pressure3pm","Temp9am","Temp3pm","RainToday"]
X =data_set[feature_cols] # Features
y = data_set.RainTomorrow # Target variable

print(X)
print(y)

```

```python
#Splitting Data
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)

#Model Development and Prediction
# import the class

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='lbfgs', max_iter=1000) 

# fit the model with data 

logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
df1=pd.DataFrame(X_test) #test data
print(df1.to_string()) #pred. data 
print(y_pred) 
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df1.to_string())
```

```python
#Evaluating the Algorithm

from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred))) 

# predicting the accuracy score

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Extracting Independent and dependent Variable

x= data_set.iloc[:, :-1].values
y= data_set.iloc[:, 5].values 
df1=pd.DataFrame(x) 
print("X=") 
print(df1.to_string())
df2=pd.DataFrame(y)
print("Y=")
print(df2.to_string())

```

```python
# Splitting the dataset into training and test set.

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,
random_state=0)
print("x_train before scaling")
df3=pd.DataFrame(x_train)
print(df3)
```

```python
#Fitting K-NN classifier to the training set

from sklearn.neighbors import KNeighborsClassifier
#n_neighbors: To define the required neighbors of the algorithm.Usually, it takes 5.
#metric='minkowski': This is the default parameter and it decides the distance between the points.
#p=2: It is equivalent to the standard Euclidean metric.
classifier=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
classifier.fit(x_train,y_train)
#Predicting the test set result
y_pred=classifier.predict(x_test)
print(y_pred)
```

```python
print("Prediction comparison")
df5=pd.DataFrame({"Y_test":y_test,"Y-pred":y_pred})
df5
```

```python
# evaluate predictions
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:%.2f"%(accuracy*100))
```
